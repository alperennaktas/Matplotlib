```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
hafta=[1,2,3,4,5]
```

```python
satıs=[10,20,30,40,80]
```

```python
numpyhafta = np.array(hafta)
numpysatıs = np.array(satıs)
```

```python
plt.subplot(1,2,1)
plt.plot(numpysatıs,numpyhafta,"r--")
plt.subplot(1,2,2)
plt.plot(numpysatıs,numpyhafta,"g*-")
```

```python
plt.plot(numpysatıs,numpyhafta)
```

```python
grafik =plt.figure()
figureAxes = grafik.add_axes([0.4,0.2,0.9,0.9])
figureAxes.plot(numpyhafta,numpysatıs,"g")
figureAxes.set_xlabel("x ekseni Veri İsmi")
figureAxes.set_ylabel("y ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```
